# 1. Remove nan data

```python
import pandas as pd
import pprint as pp
import numpy as np
```

```python
# data = pd.read_excel('data/train_set/data.xlsx', header = 1)
data = pd.read_excel('data/test_set/test_set.xlsx', header = 1)
```

```python
code_bun = data[['노출(분)','방송일시']].dropna().drop_duplicates().reset_index(drop = True)
code_bun = code_bun.set_index('방송일시')
code_bun
```

```python
nans = data.loc[(data.loc[data['노출(분)'].isnull()].index)].drop('노출(분)', axis = 1)
nans.head(3)
```

```python
nans_merged = pd.merge(nans,code_bun, how = 'left', on ='방송일시')
nans_merged.isnull().sum()
```

```python
data_remove_nan = pd.concat([data.dropna(subset=['노출(분)']), nans_merged])
data_remove_nan.isnull().sum()
data_remove_nan.sort_values(by = ['방송일시','마더코드','상품코드']).reset_index(drop = True)
```

```python
# this code is only for  train set
data_remove_nan= data_remove_nan.dropna(subset = ['취급액']).reset_index(drop = True)
data_remove_nan.isnull().sum()
```
