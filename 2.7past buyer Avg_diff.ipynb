# 2.7 Past buyer avg_diff
이전 구매자 수 와의 차이 비율 변수 생성

```python
import pandas as pd
import numpy as np
import pdb
import seaborn as sns
```

```python
data=pd.read_csv('data/train_set/2.6buyer_count_diff.csv')
data
```

```python
data['pastBuyer_count_diff']=np.nan
data['pastBuyer_count_mean']=np.nan
```

```python
data
```

```python
# code for get mean of past buyer

for i in range(len(data)):
    past_data = data.iloc[0:i]
    #find past record
    past_same_data = past_data.loc[past_data.product_code == data.loc[i].product_code] 
    if(len(past_same_data) !=0):
        pastAvg = past_same_data.buy_count.mean()
        pastAvg_diff_ = data.loc[i].buy_count/pastAvg
        data.loc[i,'pastBuyer_count_diff'] = pastAvg_diff_
        data.loc[i,'pastBuyer_count_mean']=pastAvg
```

```python
data.loc[30:60,['product_name','buy_count','buyer_count_CaAvgdiff','pastBuyer_count_mean','pastBuyer_count_diff']]
```

```python
ax = sns.boxplot(y="buyer_count_CaAvgdiff", data=data)
print(data['buyer_count_CaAvgdiff'].var())
```

```python
ax = sns.boxplot(y="pastBuyer_count_diff", data=data)
print(data['pastBuyer_count_diff'].var())
```
