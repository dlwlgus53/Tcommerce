# 2.3 업무시간/비 업무시간 변수 생성
- 휴일이 아닌날의 오전8시~ 오후 7시 사이는 일하는 시간으로 지정
- work_time변수에 기록

```python
import pandas as pd
```

```python
# data = pd.read_csv('data/train_set/2.2data_weather_watching.csv')
data = pd.read_csv('data/test_set/2.data_add_feature1.csv')
data.head()
```

```python
# 전부 일하는 시간으로 초기화
data['working_time'] = True
```

```python
# 휴일은 전부 일하지 않는 시간으로 지정
data.loc[data['off_day'] == True,'working_time'] = False
```

```python
# 9시-6시를 회사에 있는 시간으로 할 때, 8시 - 7시를 working time, 나머지 시간을 집에있는 시간으로
data.loc[data['hour'].isin([6,7,20,21,22,23,0,1,2]),'working_time'] = False
```
