## 2. add weather
- 기상청에서 다운받은 weather data
- 서울의 기온, 강수량, 강설량,습도, 부산의 기온, 강수량, 강설량, 습도
- https://data.kma.go.kr/data/grnd/selectAsosRltmList.do?pgmNo=36

```python
# weather = pd.read_csv('data/train_set/weather.csv', header = 0)
weather = pd.read_csv('data/test_set/weather.csv', header = 0)
weather.head(10)
weather.isnull().sum()
```

- 강수량과 적설의 nan을 0 으로 바꾼다.

```python
weather = weather.fillna(0)
weather
```

```python
#사용하기 편하도록 형태를 바꿔준다.
weather['일시']= pd.to_datetime(weather['일시'])
weather['date'] = weather['일시'].dt.date
weather['hour'] = weather['일시'].dt.hour
```

```python
#  기온_서울, 기온_부산 강수량_서울, 강수량_부산 이런식으로 바꿔줄 예정
weather_seoul = weather.loc[weather['지점명'] == '서울'].drop(columns= ['지점','지점명','일시'])
weather_busan = weather.loc[weather['지점명'] == '부산'].drop(columns= ['지점','지점명','일시']).reset_index(drop = True)

```

```python
weather_new =pd.merge(weather_seoul, weather_busan, suffixes=('_seoul', '_busan'), on=['date','hour'], how = 'left')
weather_new.head()
```

- 날씨데이터를 합쳐준다
- 기존 data 가 '분'도 있어서 시간단위로 만들어줘야함

```python
data['hour'] = data['방송일시'].dt.hour
data.head()
```

```python
data_weather = pd.merge(data, weather_new, how = 'left', on = ['date', 'hour'])
data_weather
```

## 3. add 미세먼지
- 미세먼지경보의 발령시간~ 해제시간 동안의 기록한다
- https://www.airkorea.or.kr/web

```python
pmWarning = pd.read_excel('data/train_set/pmWarning.xls', header =0)
pmWarning_seoul = pmWarning.loc[pmWarning['지역']=='서울']
pmWarning_busan = pmWarning.loc[pmWarning['지역']=='부산'].reset_index(drop = True)
pmWarning_busan
```

```python
bad_days_seoul = []
for i in range(len(pmWarning_seoul)):
    bad_days_seoul.extend(list(pd.date_range(end = pd.to_datetime(pmWarning_seoul['해제시간'][i]) , start = pd.to_datetime(pmWarning_seoul['발령시간'][i]), freq='H')))

bad_days_busan = []
for i in range(len(pmWarning_busan)):
    bad_days_busan.extend(list(pd.date_range(end = pd.to_datetime(pmWarning_busan['해제시간'][i]) , start = pd.to_datetime(pmWarning_busan['발령시간'][i]), freq='H')))
bad_days_busan   
```

```python
data_weather['pmWarning_seoul']= False
data_weather['pmWarning_busan']= False
```

### 서울, 부산 미세먼지 경보 추가
- 위 과정에서 정리된 미세먼지 정보를 data frame에 추가
- test set 에는 미세먼지 경보/주의보 발령기간이 이 없음

```python
for i in range(len(bad_days_seoul)):
    data_weather.loc[(data_weather['방송일시'].dt.date == bad_days_seoul[i].date()) &
                 (data_weather['방송일시'].dt.hour == bad_days_seoul[i].hour),'pmWarning_seoul'] = True
    
for i in range(len(bad_days_busan)):
    data_weather.loc[(data_weather['방송일시'].dt.date == bad_days_busan[i].date()) &
                 (data_weather['방송일시'].dt.hour == bad_days_busan[i].hour),'pmWarning_busan'] = True
```
