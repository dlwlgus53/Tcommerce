## 과거의 총 매출액 변수 생성

```python
import pandas as pd
import numpy as np
import pdb
```

```python
data=pd.read_csv('data/test_set/2.3work_time.csv')
data
```

```python
data['pastAvg_diff']=np.nan
data
```

```python
data
```

```python

for i in range(len(data)):
    past_data = data.iloc[0:i]
    past_same_data = past_data.loc[past_data.product_code == data.loc[i].product_code]
    if(len(past_same_data) ==0):
        data.loc[i,'pastAvg_diff'] = np.nan
    else:
        pastAvg = past_same_data.handling_amount.mean()
        pastAvg_diff_ = data.loc[i].handling_amount-pastAvg
        data.loc[i,'pastAvg_diff'] = pastAvg_diff_
```

```python
data.loc[0:30,['product_name','handling_amount','pastAvg_diff']]
```

```python
data.to_csv('data/train_set/2.4pastAvg_diff.csv', index = False)
```
