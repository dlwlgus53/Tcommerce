```python
data['watching_level'] = np.nan
```

```python
import pdb
def add_watching_level(df):
    pc = int(df.product_category)
    hour = int(df.hour)
    try:
        watching_level  = int(level_dict.loc[(level_dict.Product_category == pc)
                                      & (level_dict.hour == hour),'Clustering_label'])
    except : 
        watching_level  = 3
    return watching_level
    
```

```python
data['watching_level'] = data.apply(add_watching_level, axis = 1)
data
```
