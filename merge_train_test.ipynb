## 정답 제출을 위해, Train파일과 Test파일 합친 data set생성

```python
import pandas as pd
import numpy as np
```

```python
test = pd.read_csv('data/test_set/2.8_1unit_price_to_class.csv')
train = pd.read_csv('./temp2_imputation.csv')
refer = pd.read_csv('data/train_set/2.3work_time.csv')
```

```python
train['product_code'] = refer['product_code']
test['pastBuyer_count_mean']= np.nan
```

```python
for i in range(len(test)):
    #find past record
    past_same_data = train.loc[train.product_code == test.loc[i].product_code] 
    if(len(past_same_data) !=0):
        pastAvg = past_same_data.buy_count.mean()
        test.loc[i,'pastBuyer_count_mean']=pastAvg
```

```python
train['type'] = 'train'
test['type'] = 'test'

merged = pd.concat([train,test])
merged
```

```python
merged.columns
```

```python
merged = merged.drop(columns = ['handling_amount','product_code','avg_watching','max_watching','broadcast_code', 'product_name','broadcast_date','date'])
```

```python
merged['pmWarning_seoul'] = merged['pmWarning_seoul'].fillna(False)
merged['pmWarning_busan'] = merged['pmWarning_busan'].fillna(False)
```

## 이전 판매기록이 없는 경우, 카테고리별 평균 사용


```python
category_mean = pd.read_csv("category_mean.csv")
category_mean = category_mean.set_index("category")
```

```python
categories = category_mean.index
```

```python

for category in categories:
    merged.loc[(merged.product_category == category) & 
               (np.isnan(merged.pastBuyer_count_mean)),'pastBuyer_count_mean'] = category_mean.loc[category,'mean']
```

```python
merged['buy_count']
```

```python
merged['diff'] = merged['buy_count']/merged['pastBuyer_count_mean']
```

```python
merged['diff']
```

```python
merged = merged.reset_index(drop = True)
```
