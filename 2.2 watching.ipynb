## 2.2 시청률 데이터 불러오기
- test set에 대한 시청률은 존재하지 않는다.
- 그래서 시청률 데이터를 바로 사용하기보다는, 시청률 데이터를 이용해 변수를 만든뒤에 이를 적용해야한다.
- 19년도 12월 31일까지 시청률 데이터가 존재하기 때문에 기본 데이터에서 2020 1월데이터는 삭제하고 적절한 변수 만들기를 해야한다

```python
import pandas as pd
watching = pd.read_excel('data/train_set/watching.xlsx', header = 1)
# 맨 마지막 row, column은 평균값
watching
```

- 필요없는 행 제거

```python
watching= watching.drop(columns = '2019-01-01 to 2019-12-31')[:-1]
watching
```

```python
watching = pd.melt(watching, id_vars=['시간대'], var_name='time', value_name='value')
```

```python
watching['시간'] = watching['time'] +" "+ watching['시간대']
watching
```

```python
watching = watching.drop(columns = ['time','시간대'])
watching['시간'] = pd.to_datetime(watching['시간'])
watching = watching.sort_values('시간')
watching = watching.set_index(['시간'], drop = True)
```

- melt함수로 사용하기 편리해진 watching csv 정리

```python
watching.to_csv('data/train_set/watching_pivot.csv',index = False)
```

- getmax,get avg 로 방송 당, 평균, 최대 시청률 구하기

```python
from datetime import datetime, timedelta
data['broadcast_date'] = pd.to_datetime(data['broadcast_date'])
data['max_watching'] = data.apply(get_max, axis =  1)
```
